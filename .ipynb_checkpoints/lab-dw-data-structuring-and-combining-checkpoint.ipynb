```python
columns_order = ['customer', 'st', 'gender', 'education', 'customer lifetime value','income', 'monthly premium auto', 'number of open complaints','policy type', 'vehicle class', 'total claim amount']
```

## TABLA 2

```python
url2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
df_2 = pd.read_csv(url2)
df_2.columns = df_2.columns.str.lower()

df_2['customer lifetime value'] = df_2['customer lifetime value'].str.replace('%','').astype(float)
df_2['gender'] = df_2['gender'].replace({'Male':'M','female':'F'})
#df_2['gender'].value_counts()
```

```python
df_2 = df_2[columns_order]
df_2.head()
```

## TABLA 3

```python
url3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
df_3 = pd.read_csv(url3)
df_3.columns = df_3.columns.str.lower()
df_3 = df_3.rename(columns={'state': 'st'})
```

```python
df_3 = df_3[columns_order]
df_3.head()
```

## COMBINED DF

```python
df_combined = pd.concat([df_1,df_2,df_3])
```

```python
df_combined.head(10)
```

```python
df_combined = df_combined.dropna(subset = ['customer'])
```

```python
total_nan = df_combined['gender'].isna().sum()
```

```python
nan_index = df_combined['gender'].isna().index
```

```python
sum_gender = df_combined['gender'].value_counts().sum()
distribution_female = df_combined['gender'].value_counts().get('F',0) / sum_gender
distribution_male = df_combined['gender'].value_counts().get('M',0) / sum_gender

num_f = int(total_nan * distribution_female)
num_m = total_nan - num_f
```

```python
df_combined.loc[nan_index[:num_f], 'gender'] = 'F'
df_combined.loc[nan_index[num_f:num_f + num_m], 'gender'] = 'M'
```

```python
df_combined['gender'] = df_combined['gender'].ffill()
df_combined['customer lifetime value'] = df_combined['customer lifetime value'].bfill()
```

```python
df_combined['gender'] = df_combined['gender'].replace({'Male':'M', 'female':'F','Femal':'F'})
df_combined['st'] = df_combined['st'].replace({'AZ':'Arizona', 'WA':'Washington','Cali':'California'})
df_combined['education'] = df_combined['education'].replace({'Bachelors':'Bachelor'})
df_combined['number of open complaints'] = df_combined['number of open complaints'].replace({'1/0/00':'1','1/1/00':'1','1/2/00':'2','1/3/00':'3','1/4/00':'4','1/5/00':'5'})
```
